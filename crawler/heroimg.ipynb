```python
import requests
import os
```

```python
response = requests.get("http://pvp.qq.com/web201605/js/herolist.json")
```

```python
heroimg_dir = 'heroimg/'
```

```python
hero_json = response.json() 
hero_num = len(hero_json)
print(f"Total number: {hero_num}")

for hero in hero_json:
    avatar_name = hero['ename']
    save_file_name = os.path.join(heroimg_dir, f"{avatar_name}.jpg")
    avatar_url = f"http://game.gtimg.cn/images/yxzj/img201606/heroimg/{avatar_name}/{avatar_name}.jpg"
    
    print(f"{save_file_name}: {avatar_url}")
    
    if not os.path.exists(save_file_name):
        avatar_response = requests.get(avatar_url)
        
        with open(save_file_name, 'wb') as file:
            file.write(avatar_response.content)
```
